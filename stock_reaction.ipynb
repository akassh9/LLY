```python
import yfinance as yf
import pandas as pd
```

```python
lly = yf.Ticker("LLY")
snp = yf.Ticker("^GSPC")
```

```python
def acq_return(date_announced):
    start_dt = pd.to_datetime(date_announced)
    end_dt = start_dt + pd.Timedelta(days=4)

    start_str = start_dt.strftime('%Y-%m-%d')
    end_str = end_dt.strftime('%Y-%m-%d')
    
    acq = yf.Ticker("LLY").history(start=start_str, end=end_str, interval='1d')["Close"]
    mkt = yf.Ticker("^GSPC").history(start=start_str, end=end_str, interval='1d')["Close"]
    
    ret_acq = acq.pct_change().dropna()
    ret_mkt = mkt.pct_change().dropna()
    
    excess_return = ret_acq - ret_mkt
    
    return excess_return.sum() 
```

```python
user_date = input("Enter the announcement date (YYYY-MM-DD): ")

try:
    pd.to_datetime(user_date)
    result = acq_return(user_date)

    print(f"\nAnalysis for: {user_date}")
    print(f"Cumulative Abnormal Return: {result:.4%}")

except Exception as e:
    print(f"Error: Please ensure the date is in YYYY-MM-DD format. Details: {e}")
```
